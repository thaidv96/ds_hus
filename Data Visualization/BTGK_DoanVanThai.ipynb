```python
## Đọc dữ liệu vào python
import pandas as pd

```

```python
data = pd.read_csv("./dataset/train.csv")
```

```python
data.head()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
data.SalePrice.hist(bins = 50, figsize = (10,4))
plt.title("Sale Price Histogram")
plt.show()
```

```python
import numpy as np
np.log(data.SalePrice).hist(bins = 50, figsize = (10,4))
plt.title("Log(Sale Price) Histogram")
plt.show()
```

```python
sns.set_style("darkgrid")
data.plot(kind ='scatter', y = 'SalePrice', x= 'YearBuilt', figsize = (14,7))
```

```python
data.groupby('YearBuilt').SalePrice.mean().plot(figsize=(10,4))
plt.title("Mean Sale Price by Year Built")
plt.show()
```

```python

ax = data.groupby("YearBuilt").SalePrice.count().plot(kind='line',
                                    figsize=(14,8),
                                    title="Number of Houses by Year Built")
ax.set_xlabel("Year Built")
ax.set_ylabel("Frequency")
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)

ax = sns.violinplot(x="FullBath", y="SalePrice", data=data, inner = 'point', ax= ax)
sns.despine()

plt.title("Sale Price Density by # Full Bath")
plt.show()
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)

ax = sns.violinplot(x="OverallCond", y="SalePrice", data=data, inner = 'point', ax= ax)
sns.despine()

plt.title("Sale Price Density by Overall Condition")
plt.show()
```

```python
sns.set_style("darkgrid")

data.groupby("OverallCond").SalePrice.mean().plot(figsize = (14,7), kind= 'bar', color = 'blue')
plt.title("Mean Sale Price by Condition")
plt.show()
```

```python
sns.set_style('darkgrid')
data.HouseStyle.value_counts().plot(kind = 'pie', figsize = (14,7),)
plt.title("Number of house by style")
plt.show()
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)

ax = sns.violinplot(x="HouseStyle", y="SalePrice", data=data, inner = 'point', ax= ax)
sns.despine()

plt.title("Sale Price Density by HouseStyle")
plt.show()
```

```python
sns.set_style('darkgrid')
data.GarageCars.value_counts().plot(kind = 'pie', figsize = (14,7),autopct='%1.0f%%',
                                   pctdistance=1.1, labeldistance=0.5)
plt.title("Number of house by Garage Capacity")
plt.show()
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)

ax = sns.boxplot(x="GarageCars", y="SalePrice", data=data, ax= ax)

plt.title("Sale Price by Garage Capacity")
plt.show()
```
